3. Write a program that prompts the user to input a Celsius temperature and outputs the
equivalent temperature in Fahrenheit. The formula to convert the temperature is: F = 9/5 C +
32 where F is the Fahrenheit temperature and C is the Celsius temperature.

```python
C=int(input("Enter Temperature in Celsius:"))
F=(C*1.8)+32
print("Temperature in Fahrenheit is:",F)
```

4. Write a program which accept principle, rate and time from user and print the simple
interest.
The formula to calculate simple interest is: simple interest = principle * rate *
time / 100.

```python
principle=int(input("Enter Principle:"))
rate=int(input("Enter Rate:"))
time=int(input("Enter Time:"))
print("Simple Interest is:",(principle*rate*time)/100)
```

5. Write a program that accepts seconds from keyboard as integer. Your program should
converts seconds in hours, minutes and seconds.
Your output should like this :
Enter seconds: 13400
Hours: 3
Minutes: 43
Seconds: 20

```python
sec=int(input("Enter Seconds:"))
print("Hours:",sec//3600)
print("Minutes:",(sec%3600)//60)
print("Seconds:",(sec%3600)%60)
```

6. Write a program that prompts the user to enter number in two variables and swap the
contents of the variables. (Do not declare extra variable.)

```python
x,y=map(int,input("Enter two variables separated by space:").split())
print(f'Before Swapping: x={x} and y={y}')
x=x+y
y=x-y
x=x-y
#x,y=y,x
print(f'After Swapping: x={x} and y={y}')
```

9. Write a program that prompts the user to input the length and the width of a rectangle and
outputs the area and circumference of the rectangle.

```python
l=int(input("Enter Length of Rectangle:"))
w=int(input("Enter Width of Rectangle:"))
print("Area of Rectangle is:",l*w)
print("Circumference of Rectangle is:",2*(l+w))
```

10. Suppose a, b, and c denote the lengths of the sides of a triangle.
Then the area of the triangle can be calculated using the formula:
image.png
Write a program that asks the user to input the length of sides of the triangle and print the
area.

```python
a,b,c=map(int,input("Enter sides of triangle separated by space:").split())
s=(a+b+c)/2
import math
area=math.sqrt(s*(s-a)*(s-b)*(s-c))
print(f'Area={area}')
```

11. Write a program which prompts the user to input principle, rate and time and calculate
compound interest. The formula is :image.png

```python
p=int(input("Enter Principle:"))
r=int(input("Enter Rate:"))
t=int(input("Enter Time:"))
ci=p*(1+(r/100)*t)-p
print(f'Compound Interest={ci}')
```

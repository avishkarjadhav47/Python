```python
#1
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
#2
char = input("Enter a character: ")
if char.isalpha():
    print("Alphabet")
else:
    print("Not an Alphabet")
```

```python
#3
char = input("Enter a character: ")
if char.isalpha():
    if char in "aeiouAEIOU":
        print("Vowel")
    else:
        print("Consonant")
else:
    print("Not an Alphabet")
```

```python
#4
char = input("Enter a character: ")
if char.isalpha():
    print("Alphabet")
elif char.isdigit():
    print("Digit")
else:
    print("Special Character")
```

```python
#5
week_num = int(input("Enter a week number: "))
dic = {1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Saturday", 7: "Sunday"}
if week_num in dic:
    print(dic[week_num])
```

```python
#6
a,b,c=map(float,input("Enter the first angle: ").split())
if a+b+c==180:
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
#7
a,b,c=map(float,input("Enter the first side: ").split())
if a+b>c and b+c>a and a+c>b:
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
#8
a,b,c=map(float,input("Enter the first side: ").split())
if a==b==c:
    print("Equilateral Triangle")
elif a==b or b==c or a==c:
    print("Isosceles Triangle")
else:
  print("Scalene Triangle")
```

```python
#9
cost_price=float(input("Enter the cost price: "))
selling_price=float(input("Enter the selling price: "))
if selling_price>cost_price:
    print(f"Profit={selling_price-cost_price}")
elif selling_price<cost_price:
    print(f"Loss={cost_price-selling_price}")
else:
    print("No profit no loss")
```

1. Write a program to input marks of five subjects Physics, Chemistry, Biology, Mathematics and
Computers. Calculate percentage and grade according to following:
Percentage >= 90% : Grade A
Percentage >= 80% : Grade B
Percentage >= 70% : Grade C
Percentage >= 60% : Grade D
Percentage >= 40% : Grade E
Percentage < 40% : Grade F

```python
s1,s2,s3,s4,s5= map(int,input().split())
percentage= ((s1+s2+s3+s4+s5)/500)*100
print("Percentage: ",percentage)
if percentage>=90:
    print("Grade A")    
elif percentage>=80:
    print("Grade B")
elif percentage>=70:
    print("Grade C")
elif percentage>=60:
    print("Grade D")
else:    print("Grade F")        
```

2. Write a program to input basic salary of an employee and calculate its Gross salary according to
following:
Basic Salary <= 10000 : HRA = 20%, DA = 80%
Basic Salary <= 20000 : HRA = 25%, DA = 90%
Basic Salary > 20000 : HRA = 30%, DA = 95%

```python
salary= int(input("Enter your salary: "))
if salary<=10000:
    print(f'Salary:{salary+salary*0.2+salary*0.8}')
elif salary<=20000:
    print(f'Salary:{salary+salary*0.25+salary*0.9}')
elif salary>20000:
    print(f'Salary:{salary+salary*0.3+salary*0.95}')
```

3. Write a program to input electricity unit  and calculate total electricity bill according to the
given condition:
For first 50 units Rs. 0.50/unit
For next 100 units Rs. 0.75/unit
For next 100 units Rs. 1.20/unit
For unit above 250 Rs. 1.50/unit
An additional surcharge of 20% is added to the bill

```python
unit=float(input("Enter the unit: "))
if unit<=50:
    bill= unit*0.5
elif unit<=150:
    bill= 25+(unit-50)*0.75
elif unit<=250:
    bill= 100+(unit-150)*1.2
else:    bill= 220+(unit-250)*1.5
print(f'Bill: {bill}')
```

4. Write a Python program to check the validity of passwords input by users.
Validation:
a. At least 1 letter between [a-z] and 1 letter between [A-Z].
b. At least 1 number between [0-9].
c. At least 1 character from [@#$%^&+=].
d. Minimum length 6 characters.
e. Maximum length 16 characters.

```python
import re
password= input("Enter the password: ")
if (6<=len(password)<=16) and re.search("[a-z]",password) and re.search("[A-Z]",password) and re.search("[0-9]",password) and re.search("[@#$%^&+=]",password):
    print("Valid password")
else:
    print("Invalid password")
```

5. Write a Python program to convert a month name to a number of days.

Expected Output:
List of months: January, February, March, April, May, June, July, August, September,
October, November, December
Input the name of Month: February
No. of days: 28/29 days

```python
month=input("Enter the month: ") 
month= month.title()   
if month in ["January","March","May","July","August","October","December"]:
    print("31 days")
elif month in ["April","June","September","November"]:
    print("30 days")
elif month=="February":
    print("28/29 days")
    
```

6. Write a Python program that reads two integers representing a month and day and prints the season
for that month and day.
Expected Output:
Input the month (e.g. January, February etc.): july
Input the day: 31
Season is autumn

```python
month = input("Input the month (e.g. January, February etc.): ").lower()
day = int(input("Input the day: "))

if (month == "december" and day >= 21) or month in ["january", "february"] or (month == "march" and day <= 19):
    season = "winter"

elif (month == "march" and day >= 20) or month in ["april", "may"] or (month == "june" and day <= 20):
    season = "spring"

elif (month == "june" and day >= 21) or month in ["july", "august"] or (month == "september" and day <= 22):
    season = "summer"

else:
    season = "autumn"

print("Season is", season)
```

```python
#1
s = {10, 20, 30, 40}
print(s)
```

```python
#2
s = set([10, 20, 30, 40])
print(s)
```

```python
#3
s = set()
print(s)
```

```python
#4
lst = [10, 20, 30, 20, 40]
s = set(lst)
print(s)
```

```python
#5
tup = (10, 20, 30, 20, 40)
s = set(tup)
print(s)
```

```python
#6
s = {10, 20, 30}
s.add(40)
print(s)
```

```python
#7
s = {10, 20, 30}
s.update([40, 50, 60])
print(s)
```

```python
#8
s = {10, 20, 30, 40}
s.discard(20)
print(s)
```

```python
#9
s = {10, 20, 30, 40}
s.remove(20)
print(s)
```

```python
#10
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.union(s2))
```

```python
#11
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.intersection(s2))
```

```python
#13
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.difference(s2))
```

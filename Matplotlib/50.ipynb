```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Month':['Jan','Feb','Mar','Apr','May'],
    'Sales':[100,120,90,150,170],
    'Profit':[20,25,18,35,40],
    'Category':['A','B','A','B','A']
})

# Q1-Q5
plt.figure()
plt.figure(figsize=(8,5))
plt.title("Sales Chart")

# Q6-Q10
plt.plot(df['Month'],df['Sales'])
plt.plot(df['Month'],df['Sales'],marker='o')
plt.plot(df['Month'],df['Sales'],linestyle='--')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

# Q11-Q15
plt.bar(df['Month'],df['Sales'])
plt.barh(df['Month'],df['Sales'])
plt.bar(df['Month'],df['Sales'],width=0.5)

for i,v in enumerate(df['Sales']):
    plt.text(i,v,str(v))

plt.bar(df['Month'],df['Sales'])
plt.bar(df['Month'],df['Profit'])

# Q16-Q20
plt.hist(df['Sales'])
plt.hist(df['Sales'],bins=5)
plt.hist(df['Sales'],bins=[80,100,120,140,160,180])

# Q21-Q25
plt.pie(df['Sales'])
plt.pie(df['Sales'],autopct='%1.1f%%')
plt.pie(df['Sales'],explode=[0.1,0,0,0,0])
plt.pie(df['Sales'],labels=df['Month'])
plt.title("Sales Share")

# Q26-Q30
plt.scatter(df['Sales'],df['Profit'])
plt.scatter(df['Sales'],df['Profit'],s=100)
plt.scatter(df['Sales'],df['Profit'],marker='x')
plt.grid(True)

# Q31-Q35
fig,ax=plt.subplots(1,2)

fig,ax=plt.subplots(2,2)

ax[0].plot(df['Month'],df['Sales'])
ax[1].bar(df['Month'],df['Profit'])

plt.tight_layout()

# Q36-Q40
plt.figure(figsize=(12,6))
plt.plot(df['Sales'],linewidth=3)
plt.xticks(rotation=45)
plt.legend(['Sales'])
plt.savefig('chart.png')

# Q41 Highest Sales Month
plt.bar(df['Month'],df['Sales'])

# Q42 Average Sales Line
avg=df['Sales'].mean()
plt.axhline(avg)

# Q43 Sales Trend
plt.plot(df['Month'],df['Sales'])

# Q44 Profit Trend
plt.plot(df['Month'],df['Profit'])

# Q45 Sales Contribution %
plt.pie(df['Sales'],labels=df['Month'],autopct='%1.1f%%')

# Q46 Compare Categories
df.groupby('Category')['Sales'].sum().plot(kind='bar')

# Q47 Sales Distribution
plt.hist(df['Sales'])

# Q48 Cumulative Sales
plt.plot(df['Sales'].cumsum())

# Q49 Moving Average
plt.plot(df['Sales'].rolling(2).mean())

# Q50 Dual Axis
fig,ax1=plt.subplots()

ax2=ax1.twinx()

ax1.plot(df['Month'],df['Sales'])
ax2.plot(df['Month'],df['Profit'])
```

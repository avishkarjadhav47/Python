```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["Amit","Neha","Rahul","Priya","Karan"],
    "Age":[22,24,23,25,21],
    "Department":["IT","HR","IT","Finance","HR"],
    "Salary":[50000,45000,60000,70000,40000],
    "City":["Mumbai","Pune","Mumbai","Delhi","Pune"]
})

# Q1
pd.DataFrame(...)

# Q2-Q10
df.head()
df.tail(3)
df.shape
df.columns
df.dtypes
df.describe()
df.info()
len(df)
df.shape[1]

# Q11-Q20
df["Name"]
df[["Name","Salary"]]
df.iloc[0]
df.iloc[:3]
df.iloc[:,3]
df.iloc[1]
df.iloc[-1]
df[["Age","Salary"]]
df.iloc[1:4]
df.iloc[1:4,1:4]

# Q21-Q30
df[df["City"]=="Mumbai"]
df[df["Salary"]>50000]
df[df["Department"]=="IT"]
df[df["Salary"].between(45000,65000)]
df[(df["Age"]>22) & (df["City"]=="Mumbai")]
df[df["Department"].isin(["HR","IT"])]
df[df["Salary"]!=40000]
df[df["Age"]<df["Age"].mean()]
df.loc[df["Salary"].idxmax()]
df.loc[df["Salary"].idxmin()]

# Q31-Q35
df.sort_values("Salary")
df.sort_values("Salary",ascending=False)
df.sort_values(["Age","Salary"])
df.nlargest(2,"Salary")
df.loc[df["Age"].idxmin()]

# Q36-Q40
df["Salary"].mean()
df["Salary"].max()
df["Salary"].min()
df["Salary"].sum()
df["Salary"].median()

# Q41-Q45
df.groupby("Department")["Salary"].mean()
df.groupby("Department")["Salary"].max()
df.groupby("Department")["Name"].count()
df.groupby("City")["Age"].mean()
df.groupby("Department")["Salary"].sum()

# Q46-Q50
df.isnull().sum()
df.dropna()
df["Salary"].fillna(df["Salary"].mean())
df["City"].fillna(df["City"].mode()[0])
df.isnull().values.any()

# Q51-Q55
df["Bonus"]=df["Salary"]*0.10
df["Tax"]=df["Salary"]*0.05
df["Salary_After_Tax"]=df["Salary"]-df["Tax"]

df["Age_Group"]=np.where(
    df["Age"]<23,
    "Junior",
    "Senior"
)

df["Name"].str.upper()

# Q56-Q60
df[df["Name"].str.startswith("A")]
df[df["Name"].str.endswith("a")]
df[df["Name"].str.contains("h")]
df["Name"].str.lower()
df["Name"].str.len()

# Q61-Q70
df["Department"].unique()
df["Department"].value_counts()
df.duplicated()
df.drop_duplicates()

df.groupby("Department")["Salary"].max()

df.nlargest(2,"Salary").iloc[-1]

df.groupby("Department").apply(
    lambda x: x.nlargest(2,"Salary")
)

df["Salary_Rank"]=df["Salary"].rank(
    ascending=False
)

df["Dept_Rank"]=df.groupby(
    "Department"
)["Salary"].rank(
    ascending=False
)

df["Salary_Pct"]=(
    df["Salary"]/
    df["Salary"].sum()
)*100

# Q71-Q80

df2 = pd.DataFrame({
    "Department":["IT","HR","Finance"],
    "Manager":["A","B","C"]
})

pd.merge(df,df2,on="Department")
pd.merge(df,df2,on="Department",how="left")
pd.merge(df,df2,on="Department",how="inner")

pd.concat([df,df])

pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="mean"
)

pd.crosstab(
    df["Department"],
    df["City"]
)

df["Department"].value_counts()

df["Salary"].apply(lambda x:x*1.1)

df["Department"].map({
    "IT":"Tech",
    "HR":"Human Resource",
    "Finance":"Accounts"
})

df["Salary"].apply(
    lambda x:"High" if x>50000 else "Low"
)

# Q81-Q90

df[df["Name"].duplicated()]

df[df["Salary"]>
   df["Salary"].mean()]

df.nlargest(3,"Salary")

df.groupby("Department")[
    "Salary"
].mean().idxmax()

df["City"].value_counts()

df.loc[
    df.groupby("City")[
        "Salary"
    ].idxmax()
]

dept_avg = df.groupby(
    "Department"
)["Salary"].transform("mean")

df["Salary_Diff"] = (
    df["Salary"] - dept_avg
)

df["Running_Total"] = (
    df["Salary"].cumsum()
)

df["Dept_Rank"] = (
    df.groupby("Department")
      ["Salary"]
      .rank(ascending=False)
)

df.to_csv(
    "employees.csv",
    index=False
)
```

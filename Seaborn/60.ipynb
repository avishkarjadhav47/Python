```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Age':[22,25,21,30,28,26,24,29],
    'Salary':[25000,40000,22000,60000,55000,45000,30000,52000],
    'Department':['IT','HR','IT','Finance','Finance','HR','IT','Finance'],
    'Gender':['M','F','M','F','M','F','M','F']
})

# Q1-Q5
import seaborn as sns
sns.__version__
tips = sns.load_dataset('tips')
sns.set_theme()
sns.set_style('whitegrid')

# Q6-Q10
sns.scatterplot(data=df,x='Age',y='Salary')
sns.scatterplot(data=df,x='Age',y='Salary',hue='Department')
sns.scatterplot(data=df,x='Age',y='Salary',s=100)
plt.title("Age vs Salary")
sns.scatterplot(data=df,x='Age',y='Salary',hue='Department',style='Gender')

# Q11-Q15
sns.lineplot(data=df,x='Age',y='Salary')
sns.lineplot(data=df,x='Age',y='Salary',marker='o')
sns.lineplot(data=df,x='Age',y='Salary',hue='Gender')
plt.title("Line Plot")
sns.lineplot(data=df,x='Age',y='Salary',linewidth=3)

# Q16-Q20
sns.barplot(data=df,x='Department',y='Salary')
sns.barplot(data=df,x='Department',y='Salary',hue='Gender')
sns.barplot(data=df,y='Department',x='Salary')
plt.title("Department Salary")
sns.barplot(
    data=df,
    x='Department',
    y='Salary',
    order=df.groupby('Department')['Salary'].mean().sort_values().index
)

# Q21-Q25
sns.histplot(df['Salary'])
sns.histplot(df['Salary'],bins=5)
sns.histplot(df['Salary'],kde=True)
sns.histplot(df['Salary'],color='red')
plt.title("Salary Histogram")

# Q26-Q30
sns.boxplot(y=df['Salary'])
sns.boxplot(data=df,x='Department',y='Salary')
sns.boxplot(data=df,x='Gender',y='Salary')
sns.boxplot(data=df,y='Department',x='Salary')

# Q31-Q35
sns.violinplot(y=df['Salary'])
sns.violinplot(data=df,x='Department',y='Salary')
sns.violinplot(data=df,x='Gender',y='Salary')
sns.violinplot(data=df,x='Gender',y='Salary',split=False)
sns.violinplot(data=df,x='Department',y='Salary',hue='Gender')

# Q36-Q40
sns.countplot(data=df,x='Department')
sns.countplot(data=df,x='Gender')
sns.countplot(data=df,x='Department',hue='Gender')
sns.countplot(data=df,y='Department')
plt.title("Department Count")

# Q41-Q45
sns.pairplot(df)
sns.pairplot(df,hue='Gender')
sns.pairplot(df,hue='Department')
sns.pairplot(df,diag_kind='hist')
sns.pairplot(df,diag_kind='kde')

# Q46-Q50
sns.heatmap(df[['Age','Salary']].corr())
sns.heatmap(df[['Age','Salary']].corr(),annot=True)
sns.heatmap(df[['Age','Salary']].corr(),cmap='coolwarm')
sns.heatmap(df[['Age','Salary']].corr(),square=True)
plt.title("Correlation Heatmap")

# Q51-Q60
sns.barplot(data=df,x='Department',y='Salary')
sns.boxplot(data=df,x='Department',y='Salary')
sns.histplot(df['Age'],kde=True)
sns.scatterplot(data=df,x='Age',y='Salary')
sns.countplot(data=df,x='Department')
sns.countplot(data=df,x='Gender')
sns.countplot(data=df,x='Department',hue='Gender')
sns.heatmap(df[['Age','Salary']].corr(),annot=True)
sns.boxplot(data=df,x='Department',y='Salary')
```
